```python
#Gabriel Karim Miranda Carrion        Agosto 2018

import numpy as np
import scipy.integrate as integrate #Integrador
import matplotlib.pyplot as plt #Graficador

size_arr = 10000 #Tamaño del arreglo
N = np.linspace(1, 10000, size_arr)  #El rango de N

X1 =[]   #Arreglo del promedio con 10 puntos
X2 =[]   #Arreglo del promedio con 100 puntos
X3 =[]   #Arreglo del promedio con 1000 puntos
X4 =[]   #Arreglo del promedio con 10000 puntos

for i in range(100):   # 100 realizaciones de generar arreglos de números aleatorios
  
    N1_arr = np.random.randn(10) # Nos da 10 números aleatorios entre 0 y 1 con varianza 1 i.e la distribución N(0,1) 
    N2_arr = np.random.randn(100) # Nos da 100 números aleatorios entre 0 y 1 con varianza 1 i.e la distribución N(0,1) 
    N3_arr = np.random.randn(1000) # Nos da 1000 números aleatorios entre 0 y 1 con varianza 1 i.e la distribución N(0,1) 
    N4_arr = np.random.randn(10000) # Nos da 10000 números aleatorios entre 0 y 1 con varianza 1 i.e la distribución N(0,1) 
    
    X1.append(np.mean(N1_arr)) #Nos da un arreglo del promedio de un arreglo
    X2.append(np.mean(N2_arr)) #Nos da un arreglo del promedio de un arreglo
    X3.append(np.mean(N3_arr)) #Nos da un arreglo del promedio de un arreglo
    X4.append(np.mean(N4_arr)) #Nos da un arreglo del promedio de un arreglo

    
stdev1 = np.std(X1) #Desviación estándar del arreglo 1 
stdev2 = np.std(X2) #Desviación estándar del arreglo 2 
stdev3 = np.std(X3) #Desviación estándar del arreglo 3 
stdev4 = np.std(X4) #Desviación estándar del arreglo 4 
std_arr = [stdev1, stdev2, stdev3, stdev4]   #Arreglo de la desviación estándar

print std_arr

plt.figure(figsize = (12,8))     # Indica el tamaño de la grafica (x,y) 
plt.plot([10,100,1000,10000], std_arr, marker='o', markersize=5, linestyle='none' , color='b', label=r'Datos') # label will be use by plt.legend()
plt.plot(N, 1/np.sqrt(N) , linestyle='--' , linewidth=2., color='g', label=r'$\dfrac{1}{\sqrt{N}}$') # label will be use by plt.legend()
plt.xlabel('N',fontsize=20)
plt.ylabel(r'Error',fontsize=20)
plt.legend(fontsize=20, loc=1)
plt.xscale('log')
plt.show()
 
```
